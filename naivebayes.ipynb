```python
import pandas as pd
import numpy as np

x =  pd.read_csv('tabel_pesut2.csv', sep=';', header=None)



y =  pd.read_csv('tabel_x.csv', sep=';', header=None)

tes =  pd.read_csv('test_pesut.csv', sep=';', header=None)
y_tes =pd.read_csv('y_tes.csv', sep=';', header=None)

tes
```

```python
x.info()
```

```python

# Import train_test_split function
#from sklearn.model_selection import train_test_split
#x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30, random_state = 123)
#x_train

# Import Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB
# Mengaktifkan/memanggil/membuat fungsi klasifikasi Naive bayes
modelnb = GaussianNB()

```

```python
# Memasukkan data training pada fungsi klasifikasi naive bayes
nbtrain = modelnb.fit(x, y)
nbtrain.class_count_

```

```python
# Menentukan hasil prediksi dari tes
y_pred = nbtrain.predict(tes)
y_pred
```

```python
# Menentukan probabilitas hasil prediksi
y_prob = nbtrain.predict_proba(tes)
y_prob
```

```python
# import confusion_matrix model
from sklearn.metrics import confusion_matrix
confusion_matrix(y_tes, y_pred)
```
